```python
import cv2
import matplotlib.pyplot as plt
face=cv2.CascadeClassifier(r'C:\Users\HP\anaconda3\Lib\site-packages\cv2\data\haarcascade_frontalface_default.xml')
img=cv2.imread('cnnpractice.jpg')
plt.imshow(img) #BGR
```

```python
original = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(original)
```

```python
gr=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
plt.imshow(gr)
```

```python
arrays=face.detectMultiScale(gr, #input image
                             scaleFactor=1.1, # how much image size will be reduced with each scale
                             minNeighbors=2, # how many faces you want to detect
                             minSize=(120,120)) # min poosible image size
arrays
```

```python
img=cv2.imread('cnnpractice.jpg')
for (x,y,w,h) in arrays:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),5)
    
# x,y -> starting coordinate , x+w,y+h -> ending coordinate
# BGR (255,0,0) color of rectangle , 5 -> thickness of the rectangle
```
