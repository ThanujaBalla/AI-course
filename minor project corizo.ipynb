```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("spotify dataset.csv")
df
```

```python
df.isnull().sum()
```

```python
df.dropna(how='any',inplace=True)
df
```

```python
df.isnull().sum()
```

```python
# as there will be a same artist, playlist_name, playlist_genre, playlist_subgenre for multiple songs we can use label encoder
# to convert those categorical data to the numerical data. 
# in case of track_id, track-album_id, track_name they will be unique and does not effect the genre hence no need to convert them.
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

# Applying label encoder to categorical data columns
df['track_artist'] = le.fit_transform(df['track_artist'])
df['playlist_name'] = le.fit_transform(df['playlist_name'])
df['playlist_genre'] = le.fit_transform(df['playlist_genre'])
df['playlist_subgenre'] = le.fit_transform(df['playlist_subgenre'])

df.head()
```

```python
# to normalize the data we can use standard scaler
from sklearn.preprocessing import StandardScaler
numerical_features = ['track_popularity', 'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms']
scaler = StandardScaler()
df[numerical_features] = scaler.fit_transform(df[numerical_features])

df.head()
```

```python
#to perform k means clustering we need to know how many  clusters we need to make and for this we use elbow method
# the inertia of all the no of clusters are plotted and the numner where we can observe turn in the plot describes best number of clusters
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Selecting features for clustering
features = df[["track_popularity","playlist_genre","playlist_subgenre","danceability","energy","key","loudness","mode","speechiness","acousticness","instrumentalness","liveness" ,"valence" ,"tempo" ,"duration_ms"]]

# good cluster point is with low inertia and less no of clusters
# sse defines sum of squared distances
sse = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,init='k-means++',  n_init=10, random_state=42)
    kmeans.fit(features)
    sse.append(kmeans.inertia_)

# Plot the elbow graph
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), sse, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('sum of squared distances')
plt.show()
```

```python
# as we know the low inertia point is best suited for no of clusters and that should be the minimum number possible
# hence the better lowest point is 4. even the numbers after 4 are less there is no much difference in their sse and 4 is least
kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=42)
df['cluster'] = kmeans.fit_predict(features)
```

```python
df # we can see the last column which is describing the cluster that the song belongs to i.e. genre 
```

```python
# to visualize the clusters with various pairs as x and y axis
# Define a function to create scatter plots
def plot_clusters(feature_x, feature_y, x_label, y_label):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=feature_x, y=feature_y, hue='cluster', data=df, palette='viridis')
    plt.title('Clusters of Songs')
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.legend()
    plt.show()

```

```python
# Valence vs. Energy
plot_clusters('valence', 'energy', 'Valence', 'Energy')
```

```python
# Danceability vs. Tempo
plot_clusters('danceability', 'tempo', 'Danceability', 'Tempo')
```

```python
# Loudness vs. Acousticness
plot_clusters('loudness', 'acousticness', 'Loudness', 'Acousticness')
```

```python
# Speechiness vs. Instrumentalness
plot_clusters('speechiness', 'instrumentalness', 'Speechiness', 'Instrumentalness')
```

```python
# Liveness vs. Duration_ms
plot_clusters('liveness', 'duration_ms', 'Liveness', 'Duration (ms)')
```

```python
# to get the index of the 
features_index = cluster_df.index.get_loc(input_index)
features_index
```

```python
def get_nearest_tracks(df,input_track_id,n_neighbors=5):
    input_cluster=df.loc[df['track_id']==input_track_id,'cluster'].values[0]
    cluster_df = df[df['cluster'] == input_cluster]
    features = cluster_df.iloc[:,9:].reset_index(drop=True)
    input_index = cluster_df[cluster_df['track_id'] == input_track_id].index[0]
    knn = NearestNeighbors(n_neighbors=n_neighbors + 1)  # +1 because the input track will be the nearest
    knn.fit(features)
    features_index = cluster_df.index.get_loc(input_index)
    distances, indices = knn.kneighbors(features.iloc[[features_index]])
    nearest_indices = indices[0][1:]
    
    # Get the nearest track IDs
    nearest_track_ids = cluster_df.iloc[nearest_indices]['track_id'].values
    recommended_songs = []
    for track_id in nearest_track_ids:
        song_name = cluster_df[cluster_df['track_id'] == track_id]['track_name'].iloc[0]
        recommended_songs.append(song_name)
    
    return recommended_songs
```

```python
input_track_id = '6f807x0ima9a1j3VPbc7VN'
print("The current song :",df[df['track_id']==input_track_id]["track_name"].iloc[0])
print("The recommended songs: ")
nearest_tracks = get_nearest_tracks(df, input_track_id)
print(*nearest_tracks,sep="\n")
```
