```python
#series
l=[1,2,3,4,5]
series=pd.Series(l)
print(series)
print(type(series))
#by default index will be from 0
print(series[2]) 
```

```python
# to customise index 
series=pd.Series(l,index=['a','b','c','d','e'])
print(series)
print(series['b'])
```

```python
temp={'day1':32,'day2':45,'day3':42}
print(temp)
print(type(temp))

temp_series=pd.Series(temp)
print(temp_series)
print(type(temp_series))
```

```python
#to change the datatype
newtemp_series=pd.Series(temp,dtype="float32")
print(newtemp_series)
print(type(newtemp_series))
```

```python
# joining the tables ( join in SQL = Merge in pandas)
orders=pd.DataFrame({"Order_id":[1,2,3],"p_id":["A012","f342","c654"]})
orders
```

```python
products=pd.DataFrame({"p_id":["A012","f342","d546"],"c_id":[213,432,433],"p_name":["Shirts","pants","skirts"]})
products
```

```python
#inner join 
pd.merge(orders,products,how="inner",indicator=True)
# indicator will indicate from which table it is merged
```

```python
#left join
pd.merge(orders,products,how="left", indicator=True)
```

```python
#right join
pd.merge(orders,products,how="right")
```

```python
#full join
pd.merge(orders,products,how="outer",indicator=True)
```

```python
pd.concat((orders,products),axis=0) #adding table below other table
```

```python
pd.concat((orders,products),axis=1) #adding table beside other table
```
