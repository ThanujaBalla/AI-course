```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
p_id=[1,2,3,4,5]
sales=[100,234,32,32,43]
plt.plot(p_id,sales)
```

```python
plt.scatter(p_id,sales)
```

```python
sales_df=pd.DataFrame({"product_id":p_id,"Sales Amount":sales})
sales_df
```

```python
plt.plot(sales_df["product_id"],sales_df["Sales Amount"])
```

```python
plt.plot(sales_df["product_id"],sales_df["Sales Amount"])
plt.xlabel("Product Id")
plt.ylabel("Sales")
plt.title("Sales Graph")
```

```python
colours=['r','b','g','y','m']
plt.scatter(sales_df["product_id"],sales_df["Sales Amount"],marker="*",c=colours)
plt.xlabel("Product Id")
plt.ylabel("Sales")
plt.title("Sales Graph")
```

```python
colours=['r','b','g','y','m'] # changing markers colours
sizes=[100,40,50,35,60] # changing markers size
plt.scatter(sales_df["product_id"],sales_df["Sales Amount"],marker="*",c=colours,s=sizes)
plt.xlabel("Product Id")
plt.ylabel("Sales")
plt.title("Sales Graph")
```

```python
sns.lineplot(x="product_id", y="Sales Amount", data=sales_df)
```

```python
plt.bar(sales_df["product_id"],sales_df["Sales Amount"],color=colours)
plt.xlabel("Product Id")
plt.ylabel("Sales")
plt.title("Sales Graph")
```

```python
plt.bar(sales_df["product_id"],sales_df["Sales Amount"],color=colours,width=0.5)
plt.xlabel("Product Id")
plt.ylabel("Sales")
plt.title("Sales Graph")
```

```python
plt.bar(sales_df["product_id"],sales_df["Sales Amount"],color=colours,alpha=0.5) # alpha to control colour brightness
plt.xlabel("Product Id")
plt.ylabel("Sales")
plt.title("Sales Graph")
```

```python
plt.barh(sales_df["product_id"],sales_df["Sales Amount"],color=colours)  # barh will give horizontal bar graph
plt.xlabel("Product Id")
plt.ylabel("Sales")
plt.title("Sales Graph")
```

```python
# sub plot to plot multiple plots
l1=[23,58,60,45]
l2=[67,32,45,65]
plt.plot(l1,l2)
l3=[35,56,43,67]
l4=[55,45,32,34]
plt.plot(l3,l4)
# we are getting two plots in same graph
```

```python
# to plot them in two different plots
l1=[23,58,60,45]
l2=[67,32,45,65]
plt.subplot(1,2,1) # no of rows, no of columns, position of the graph
plt.plot(l1,l2)
plt.title("l1 vs l2")
plt.xlabel("l1")
plt.ylabel("l2")

l3=[35,56,43,67]
l4=[55,45,32,34]
plt.subplot(1,2,2)
plt.plot(l3,l4)
plt.title("l3 vs l4")
```

```python
# to plot them in two different plots
l1=[23,58,60,45]
l2=[67,32,45,65]
plt.subplot(2,1,1) # no of rows, no of columns, position of the graph
plt.plot(l1,l2)
plt.title("l1 vs l2")
plt.xlabel("l1")
plt.ylabel("l2")

l3=[35,56,43,67]
l4=[55,45,32,34]
plt.subplot(2,1,2)
plt.plot(l3,l4)
plt.title("l3 vs l4")
```
